```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(r"C:\Users\Aakash\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("https://www.cardekho.com/")
driver.maximize_window()
Search = driver.find_element(By.XPATH,'/html/body/div[2]/div/header/div[2]/div/div/nav/ul/li[2]/span/span')
Search.click()
```

```python
Brand = []
driven_kilometers = []
Model = []
Year = []
Fuel = []

Brand_name = driver.find_elements(By.XPATH,'//h3[@class="title"]')

for time in range(0,300):
    for names in Brand_name:
        n = names.text.split(' ')[1]
        Brand.append(n)
    driver.execute_script("window.scrollTo(0,300);")
    

Kilometers = driver.find_elements(By.XPATH,'//div[@class="dotsDetails"]')

for time in range(0,300):
    for kms in Kilometers:
        kiloms = kms.text.split(' ')[0]
        driven_kilometers.append(kiloms)
    driver.execute_script("window.scrollTo(0,300);")


Model_name = driver.find_elements(By.XPATH,'//h3[@class="title"]')

for time in range(0,300):
    for name in Model_name:
        models = name.text.split(' ')[2]
        Model.append(models)
    driver.execute_script("window.scrollTo(0,300);")
    

Manu_Year = driver.find_elements(By.XPATH,'//h3[@class="title"]')

for time in range(0,300):
    for year in Manu_Year:
        ManuYear = year.text.split(' ')[0]
        Year.append(ManuYear)
    driver.execute_script("window.scrollTo(0,300);")
   
print(len(Brand),len(driven_kilometers),len(Model),len(Year))
```

```python
SAles_price = []

price = driver.find_elements(By.XPATH,'//div[@class="Price hover"]//p')

for time in range(0,300):
    for prices in price:
        p = prices.text.split(' ')[1].split(' ')[0]
        SAles_price.append(p)
    driver.execute_script("window.scrollTo(0,300);")
    
print(len(SAles_price))
```

```python
Fuel1=[]
fuel_type = driver.find_elements(By.XPATH,'//div[@class="dotsDetails"]')

for time in range(0,300):
    for types in fuel_type:
        fuels = types.text.split(' ')[3]
        Fuel1.append(fuels)
    driver.execute_script("window.scrollTo(0,300);")
len(Fuel1)
Fuel1
```

```python
#DataFrame

df = pd.DataFrame({'Brand':Brand,'Kilometers':driven_kilometers,'Model_name':Model,'Manu_Year':Year,'fuel_type':Fuel1,'SAles_price':SAles_price})
df
```

```python
df.to_csv('CarPriceData.csv')
```
